```python

zone_counts = df['Zone'].value_counts()
zone_counts.plot(kind='bar', title='Number of Entries per Zone', color='lavender')
plt.xlabel('Zone')
plt.ylabel('Count')
plt.show()
```

```python
status_counts = df['Status(Fiberized/ Non Fiberize)'].value_counts()
status_counts.plot(kind='pie', autopct='%1.1f%%', title='Fiberized vs. Non-Fiberized', colors=['pink', 'purple'])
plt.ylabel('')
plt.show()
```

```python
zone_status = df.groupby(['Zone', 'Status(Fiberized/ Non Fiberize)']).size().unstack()
```

```python
zone_status.plot(kind='bar', stacked=True, title='Fiberized Status by Zone', color=['purple','pink'])
plt.xlabel('Zone')
plt.ylabel('Count')
plt.show()
```

```python
carry_forward_counts = df['Carry Forwarded From 2023'].value_counts()
carry_forward_counts.plot(kind='bar', title='Carry Forwarded Data Distribution', color='grey')
plt.xlabel('Carry Forwarded From 2023')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show()


```

```python

line_data = pd.crosstab(df['City'], df['Status(Fiberized/ Non Fiberize)'])
line_data.plot(kind='line', figsize=(15,7),marker='o')
plt.title('City vs. Fiberized Status')
plt.xlabel('City')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(True)
plt.legend(title='Status (Fiberized/ Non Fiberize)')
plt.show()
```
